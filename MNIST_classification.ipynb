```python
from load_mnist import load_mnist
from main_functions import *
```

```python
no_layer = [784,10] # array of neurons when no hidden layer
single_layer = [784,128,10] # array of neurons with 1 hidden layer
several_layers = [784,128,64,64,10] # array of neurons with 3 hidden layers
k = 10
batch_size = 64
activation = "relu"
np.random.seed(seed=42)
```

```python
X_train, Y_train, X_test, Y_test = load_mnist()
```

```python
# show the shapes of train and test data
print(f'''Shape of X_train: {X_train.shape}
Shape of Y_train: {Y_train.shape}
Shape of X_test: {X_test.shape}
Shape of Y_test: {Y_test.shape}''')
```

#### Train the model with no hidden layers

```python
no_hid_train_costs, no_hid_train_accuracies, no_hid_test_costs, no_hid_test_accuracies, no_hid_batch_costs, no_hid_batch_accuracies, no_hid_state = train_model(X_train, Y_train, no_layer, 250, 0.01, batch_size, X_test, Y_test, k, activation)
```

```python
# convert values from numpy.float64 to float
no_hid_train_accuracies_float = [float(i)*100 for i in no_hid_train_accuracies]
no_hid_test_accuracies_float = [float(i)*100 for i in no_hid_test_accuracies]
```

```python
training_curve_plot('Losses and accuracies on the model with no hidden layer (Iterations: 250, Learning rate: 0.01)', no_hid_train_costs, no_hid_test_costs, no_hid_train_accuracies_float, no_hid_test_accuracies_float,250)
```

```python
# show the weights of the model with no hidden layer
show_weights(no_hid_state)
```

#### Train the model with one hidden layer

```python
one_hid_train_costs, one_hid_train_accuracies, one_hid_test_costs, one_hid_test_accuracies, one_hid_batch_costs, one_hid_batch_accuracies,one_hid_state = train_model(X_train, Y_train, single_layer, 200, 0.01, batch_size, X_test, Y_test, k, activation)
```

```python
one_hid_train_accuracies_float = [float(i)*100 for i in one_hid_train_accuracies]
one_hid_test_accuracies_float = [float(i)*100 for i in one_hid_test_accuracies]
```

```python
training_curve_plot('Losses and accuracies on the model with one hidden layer (Iterations: 200, Learning rate: 0.01)', one_hid_train_costs, one_hid_test_costs, one_hid_train_accuracies_float, one_hid_test_accuracies_float, 200)
```
